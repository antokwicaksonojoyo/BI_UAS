# 3A #perlakuan dengan kondisi Ibu hamil dengan Usia 25 Tahun, yangmerupakan Kelahiran ke -1, dengan Waktu kelahiran sesuai dengan HPL, Memiliki tekanan darah Normal


```python
dis = []
for i in range(80):
    dis.append(math.sqrt((float(df.iloc[i]['Usia'])-25)**2+
                         (float(df.iloc[i]['Kelahiran_ke-'])- 1)**2+
                         (float(df.iloc[i]['Waktu_Kelahiran'])-0)**2+
                         (float(df.iloc[i]['Tekanan_darah'])-1)**2))
```

```python
df['dis'] = dis
df
```

```python
df.sort_values('dis')
```

```python
x = df.sort_values('dis').head(5)
x
```

```python
y = x["Caesarian"]
y
```

```python
np.mean(y)
```
